## Các đặc trưng Object

```python
def check_exact_data_types(df, columns_to_check):
    """
    Hàm kiểm tra và thống kê chính xác các kiểu dữ liệu thực sự
    nằm bên trong các cột dạng 'object' của DataFrame.
    """
    for col in columns_to_check:
        # Kiểm tra xem cột có tồn tại trong DataFrame không
        if col not in df.columns:
            print(f"⚠️ Cột '{col}' không tồn tại trong DataFrame.\n")
            continue

        print(f"=== Thống kê kiểu dữ liệu cột: '{col}' ===")

        # 1. Đếm số lượng giá trị thiếu (NaN/Null)
        missing_count = df[col].isnull().sum()
        total_rows = len(df)

        # 2. Lấy dữ liệu hợp lệ (bỏ qua NaN/Null để không bị lỗi khi check type)
        valid_data = df[col].dropna()

        if valid_data.empty:
            print(f"Cột này hoàn toàn trống (toàn bộ là Null/NaN).\n")
            continue

        # 3. Lấy tên kiểu dữ liệu của từng phần tử (ví dụ: 'str', 'list', 'dict', 'ndarray')
        # Dùng thuộc tính __name__ để lấy tên kiểu dữ liệu dưới dạng chuỗi, dễ gom nhóm hơn
        type_series = valid_data.apply(lambda x: type(x).__name__)

        # 4. Đếm tần suất xuất hiện của từng kiểu
        type_counts = type_series.value_counts()

        # 5. In kết quả
        print("Chi tiết các kiểu dữ liệu đang tồn tại:")
        for dtype, count in type_counts.items():
            # Tính tỷ lệ trên tổng số dòng không bị Null
            percentage = (count / len(valid_data)) * 100
            print(f"  • {dtype:<10} : {count:>6} dòng ({percentage:.2f}%)")

        if missing_count > 0:
            print(f"  • {'Null/NaN':<10} : {missing_count:>6} dòng (Giá trị bị thiếu)")

        print("-" * 40 + "\n")

# --- CÁCH SỬ DỤNG ---
# Giả sử tên DataFrame của bạn là df
# Khai báo danh sách các cột bạn muốn kiểm tra
cols_list = ['categories', 'details', 'features', 'description', 'title_x','title_y', 'text','store']

# Chạy hàm
check_exact_data_types(df, cols_list)
```

```python
def get_empty_percentage(df, column_name):
    if column_name not in df.columns:
        return 0.0

    def is_truly_empty(val):
        if isinstance(val, (list, dict, tuple, set, np.ndarray)):
            return len(val) == 0

        # 2. Kiểm tra chuỗi
        if isinstance(val, str):
            cleaned_str = val.strip()
            if cleaned_str == "" or cleaned_str == "[]" or cleaned_str == "{}":
                return True
            return False

        try:
            if pd.isna(val):
                return True
        except ValueError:
             return False

        return False

    total_empty = df[column_name].apply(is_truly_empty).sum()
    total_rows = len(df)

    if total_rows == 0:
        return 0.0

    return (total_empty / total_rows) * 100
cols_list = ['categories', 'details', 'features', 'description', 'title_x', 'title_y', 'text', 'store']
for col in cols_list:
    ty_le_rong = get_empty_percentage(df, col)
    print(f"{col} : {ty_le_rong:.2f}%")
```

```python
def get_details_keys_df(df, column='details', min_count=1):
    if column not in df.columns:
        return pd.DataFrame()
    valid = df[column].dropna()
    valid = valid[valid.apply(lambda x: isinstance(x, dict))]
    counter = Counter()
    for d in valid:
        counter.update(d.keys())
    keys_df = pd.DataFrame(counter.items(), columns=['key', 'count'])
    keys_df = keys_df[keys_df['count'] >= min_count].sort_values('count', ascending=False)
    return keys_df

keys_df = get_details_keys_df(df, min_count=10)
print(keys_df)
```

```python
def is_non_empty_value(val):
    """Kiểm tra xem giá trị có thực sự chứa thông tin hay không (không rỗng)"""
    if pd.isna(val):
        return False
    if isinstance(val, (list, dict, tuple, set)):
        return len(val) > 0
    if isinstance(val, str):
        return val.strip() != ''
    return True
def list_top_non_empty_keys(df, column='details', top_k=10):
    """
    Liệt kê top K key trong cột details mà có giá trị không rỗng (có nội dung thực sự).
    """
    if column not in df.columns:
        print(f"Cột '{column}' không tồn tại.")
        return {}

    valid_details = df[column].dropna()
    valid_details = valid_details[valid_details.apply(lambda x: isinstance(x, dict))]

    if len(valid_details) == 0:
        print("Không có dòng nào có details hợp lệ.")
        return {}

    key_counter = Counter()
    for d in valid_details:
        for k, v in d.items():
            if is_non_empty_value(v):
                key_counter[k] += 1

    top_keys = key_counter.most_common(top_k)

    print(f"Top {top_k} key có giá trị không rỗng (có nội dung thực sự):")
    for key, count in top_keys:
        print(f"  {key}: xuất hiện {count} lần")

    return dict(top_keys)

# Sử dụng
top_keys = list_top_non_empty_keys(df, column='details', top_k=10)
```

```python
# Giả sử bạn đã có cột 'brand' hoặc trích xuất từ details
if 'brand' not in df.columns:
    df['brand'] = df['details'].apply(lambda x: x.get('Brand') if isinstance(x, dict) else None)

# Hàm xử lý rỗng (bao gồm None, NaN, chuỗi rỗng, khoảng trắng)
def clean_brand(val):
    if pd.isna(val):
        return 'unknown'
    if isinstance(val, str):
        if val.strip() == '':
            return 'unknown'
    return val

df['brand_clean'] = df['brand'].apply(clean_brand)

# Đếm số lượng brand duy nhất (không tính 'unknown')
n_brands = df[df['brand_clean'] != 'unknown']['brand_clean'].nunique()
print(f"Số lượng brand (không kể 'unknown'): {n_brands}")

# Xem top 10 brand phổ biến nhất
top_brands = df['brand_clean'].value_counts().head(10)
print("\nTop 10 brand xuất hiện nhiều nhất:")
print(top_brands)
```
